## 1. Plotting with Matplotlib

### Line plots

Let's plot the function $\displaystyle y=\frac{1-5x+7x^3}{4x^4-x+1}$ for $x \in [-2,2]$.

```python
x = np.linspace(-2,2,100)
y = (1 - 5*x + 7*x**3) / (4*x**4 - x + 1)
plt.plot(x,y)
plt.ylim([-3,3]), plt.xlim([-2,2])
plt.show()
```

### Parametric plots

To create a parametric plot, we create an array of $t$ values and then create arrays of $x$ and $y$ values.

Let's plot the [Trefoil knot](https://en.wikipedia.org/wiki/Trefoil_knot) in 2D:

$$
x(t) = \sin t + 2 \sin 2t \ \ , \ \ y(t) = \cos t - 2 \cos 2t
$$

for $t \in [0,2\pi]$. Let's also save the figure as a .png file.

```python
t = np.linspace(0,2*np.pi,100)
x = np.sin(t) + 2 * np.sin(2*t)
y = np.cos(t) - 2 * np.cos(2*t)
plt.plot(x,y)
plt.title('2D Trefoil Knot')
plt.axis('equal')
plt.axis('off')
plt.savefig('trefoil.png',dpi=200,format='png')
plt.show()
```

Let's do another example. Let's plot the heart curve:

$$
\begin{align*}
x &= 16 \sin^3(t) \\
y &= 13 \cos(t) - 5 \cos(2t) - 2 \cos(3t) - \cos(4t)
\end{align*}
$$

```python
t = np.linspace(0,2*np.pi,100)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.plot(x,y,'r',lw=20)
plt.title('Heart!')
plt.axis('equal')
plt.axis('off')
plt.show()
```

Let's plot the buterfly curve:

$$
\begin{align*}
x &= \sin(t) e^{\cos t } - 2 \cos(4t) - \sin^5(t/12) \\
y &= \cos(t) e^{\cos t } - 2 \cos(4t) - \sin^5(t/12)
\end{align*}
$$

```python
t = np.linspace(0,100,2000)
x = np.sin(t)*(np.exp(np.cos(t)) - 2*np.cos(4*t) - (np.sin(t/12))**5)
y = np.cos(t)*(np.exp(np.cos(t)) - 2*np.cos(4*t) - (np.sin(t/12))**5)
plt.plot(x,y)
plt.title('Butterfly')
plt.axis('equal')
plt.axis('off')
plt.show()
```

### Scatter plots

A scatter plot has 4 dimensions: $x$ coordinate, $y$ coordinate, size and color. Let's make a random scatter plot:

```python
N = 2000 # Number of dots in the plot
x = np.random.rand(N) # Random x coordinates sampled uniformly from [0,1]
y = np.random.rand(N) # Random y coordinates sampled uniformly from [0,1]
size = 100*np.random.rand(N) + 20 # Random size of dots sampled uniformly from [20,120]
colors = np.random.rand(N,4) # Random color 4-tuples (red,green,blue,alpha) sampled uniformly from [0,1]
plt.figure(figsize=(12,5)) # Create a figure of size 12 by 5 
plt.scatter(x,y,c=colors,s=size)
plt.axis('off')
plt.show()
```

### Histograms

Let's generate an array of 10000 random number sampled from the normal distribution and create a histogram:
